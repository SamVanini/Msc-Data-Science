# Plotting real data

The dataset used here contains Average Annual Gasoline (Petrol) Retail Prices in Selected Countries [$US per gallon]

## Imports section

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Loadind data set

```python
df = pd.read_csv('https://raw.githubusercontent.com/resbaz/lessons/refs/heads/master/matlab/intermediate/Exercises/gasprices.csv', skiprows=4)
```

## Exercise

Plot all prices in one plot with proper labeling, titles, etc.

```python
plt.figure(figsize=(10, 6))
plt.plot(df.Year, df.Australia, label="Australia")
plt.plot(df.Year, df.Canada, label="Canada")
plt.plot(df.Year, df.France, label="France")
plt.plot(df.Year, df.Germany, label="Germany")
plt.plot(df.Year, df.Italy, label="Italy")
plt.plot(df.Year, df.Japan, label="Japan")
plt.plot(df.Year, df.Mexico, label="Mexico")
plt.plot(df.Year, df['South Korea'], label="South Korea")
plt.plot(df.Year, df.UK, label="UK")
plt.plot(df.Year, df.USA, label="USA")
plt.xticks(df.Year[::3], rotation=45)
plt.title('Average Annual Gasoline (Petrol) Retail Prices')
plt.xlabel('Years')
plt.ylabel('Average Price (USD)')
plt.legend()
plt.show()
```

## Exercise

Plot the histograms of all prices in different subplots

```python
fig, axes= plt.subplots(2, 5, figsize=(30, 10), sharex=True, sharey=True)

fig.suptitle('Average Annual Gasoline (Petrol) Retail Prices')
fig.supxlabel('Years')
fig.supylabel('Average Price (USD)')
fig.tight_layout(pad=2.25)

axes[0, 0].plot(df.Year, df.Australia)
axes[0, 0].set_title('Australia')

axes[0, 1].plot(df.Year, df.Canada)
axes[0, 1].set_title('Canada')

axes[0, 2].plot(df.Year, df.France)
axes[0, 2].set_title('France')

axes[0, 3].plot(df.Year, df.Germany)
axes[0, 3].set_title('Germany')

axes[0, 4].plot(df.Year, df.Italy)
axes[0, 4].set_title('Italy')

axes[1, 0].plot(df.Year, df.Japan)
axes[1, 0].set_title('Japan')

axes[1, 1].plot(df.Year, df.Mexico)
axes[1, 1].set_title('Mexico')

axes[1, 2].plot(df.Year, df['South Korea'])
axes[1, 2].set_title('South Korea')

axes[1, 3].plot(df.Year, df.UK)
axes[1, 3].set_title('UK')

axes[1, 4].plot(df.Year, df.USA)
axes[1, 4].set_title('USA')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in fig.get_axes():
    ax.label_outer()
    ax.set_xticks(df.Year[::3])
```
